# Analyzing Museum of Modern Art data

Let's analyze the data provided by Museum of Modern Art (MoMA). The data can be found on <a href="https://github.com/MuseumofModernArt/collection">Github</a>.

```python
import pandas
import re

artworks = pandas.read_csv("Artworks.csv")
artworks.head(8)
```

The Github page mentioned that this dataset is still a work in progress so it could have some data quality issues. Printing the first 10 rows above shows that most of the columns have proper data except for "Date". Let's dig deeper into this.

```python
#artworks['Date'].count()

# http://stackoverflow.com/questions/29530232/python-pandas-check-if-any-value-is-nan-in-dataframe
artworks['Date'].isnull().sum()
```

There are various kind of corner cases with the dates. So let's start by fixing a few patterns.


```python
def clean_dates(row):
    initial_date = row['Date']
    final_date = initial_date

    # Instead of finding various patterns, just give me the first four digits
    #^[0-9]{4}$
    #final_date = re.sub(r'^[0-9]{4}$', '^[0-9]{4}$', initial_date)
    

    if initial_date:
        final_date = re.findall(r'[0-9]{4}', initial_date) #r'^[0-9]{4}$
        #http://stackoverflow.com/questions/19252588/how-do-i-test-for-null-list-entry-in-python-list
        if None in final_date:
            return None
        else:
            return final_date[0]
    else:
        return None

    
'''
for i in range(400,500):
    initial_date = artworks['Date'][i]
    #print(initial_date)
    
    final_date = initial_date
    
    if initial_date:
        final_date = re.findall(r'[0-9]{4}', initial_date)
    else:
        print("None")
        
    print("Initial_date: " + initial_date + " -> " + final_date[0])
'''  

# Assign the results of "clean_split_dates" to the 'Date' column. Since we want Pandas to go
# row by row, we set "axis=1". If we wanted to go by columns, we would use "axis=0".
artworks['Date'] = artworks.apply(lambda row: clean_dates(row), axis=1)
artworks['Date'].value_counts()

```
